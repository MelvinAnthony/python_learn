# Star Pattern

```python
n = int(input("ENter the number: "))

for i in range(1,n+1):
    print("* "*i,end="\n")
```

```python
n = 5

for i in range(1,n+1):
    space = n - i
    print("  "*space + "* "*i)
```

```python
n = 10
for i in range(1,n+1):
    space = n - i
    print(" "*space + "*"*(2*i-1))
```

```python
n = 5

# Top

for i in range(1,n+1):
    space = n - i
    print(" " * space + " *"*(2*i-1))

# Bottom
for i in range(n-1,0,-1):
    print(" " * space + "* "*(2*i-1))
```
